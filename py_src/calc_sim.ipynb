```python
### import raw data

import numpy as np
import pandas as pd

df = pd.read_excel('raw_data/refined_data.xlsx')
print("===== Summary =====")
print(df)
```

```python
### 코사인 유사도

from sklearn.metrics.pairwise import cosine_similarity

# DataFrame to NumPy array
data = df.iloc[:, 1:].values

# calculation cosine similarity about all items
cosine_sim_matrix = cosine_similarity(data)

# declare a list to save data
similarity_results = []

# save result of cosine similarity (i != j)
for i in range(len(cosine_sim_matrix)):
    for j in range(i + 1, len(cosine_sim_matrix)):
        similarity_results.append([i + 1, j + 1, cosine_sim_matrix[i, j]])

# pandas DataFrame
similarity_df = pd.DataFrame(similarity_results, columns=["Row1", "Row2", "Cosine Similarity"])

# save CSV file
similarity_df.to_csv('result/cosine_similarity_results.csv', index=False)

print("Data saved as a CSV file.")
```

```python
### 유클리드 거리 각 factor마다 비교 test1

from sklearn.metrics.pairwise import euclidean_distances

# Single column from DataFrame as a NumPy array
data_single_column = df.iloc[:, 1].values.reshape(-1, 1)

# Calculate Euclidean distance for a single column
euclidean_dist_matrix = euclidean_distances(data_single_column)

euclidean_results = []

# Display the results
for i in range(len(euclidean_dist_matrix)):
    for j in range(i + 1, len(euclidean_dist_matrix)):
        # print(f"Euclidean distance between row {i + 1} and row {j + 1}: {round(euclidean_dist_matrix[i, j], 2)}")
        euclidean_results.append([i + 1, j + 1, euclidean_dist_matrix[i, j]])

euclidean_df = pd.DataFrame(euclidean_results, columns = ["Row1", "Row2", "EXT"])
print(euclidean_df)
```
